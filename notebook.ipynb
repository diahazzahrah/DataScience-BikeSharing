## Data Wrangling

### Gathering Data

Dataset from : https://www.kaggle.com/datasets/lakshmi25npathi/bike-sharing-dataset

Attribute Information:

---
Both hour.csv and day.csv have the following fields, except hr which is not available in day.csv
*   instant: record index
*   dteday : date
*   season : season (1:springer, 2:summer, 3:fall, 4:winter)
*   yr : year (0: 2011, 1:2012)
*   mnth : month ( 1 to 12)
*   hr : hour (0 to 23)
*   holiday : weather day is holiday or not (extracted from [Web Link])
*   weekday : day of the week
*   workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
*   weathersit :
    1.   Clear, Few clouds, Partly cloudy, Partly cloudy
    2.   Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    3.   Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    4.   Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
*   temp : Normalized temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (only in hourly scale)
*   atemp: Normalized feeling temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (only in hourly scale)
*   hum: Normalized humidity. The values are divided to 100 (max)
*   windspeed: Normalized wind speed. The values are divided to 67 (max)
*   casual: count of casual users
*   registered: count of registered users
*   cnt: count of total rental bikes including both casual and registered

### Cleaning Data

```python
# Rename column , yr, mnth, hr, cnt

df.rename(columns = {'yr': 'year'}, inplace = True)
df.rename(columns = {'mnth': 'month'}, inplace = True)
df.rename(columns = {'hr': 'hour'}, inplace = True)
df.rename(columns = {'cnt': 'count'}, inplace = True)
```

Hal ini dengan tujuan memudahkan dalam melakukan proses analisis data

```python
# Mengganti tipe data pada variabel dteday
df['dteday'] = pd.to_datetime(df['dteday'])
```

```python
df.info()
```

Setelah mengatasi variabel dteday dari tipe data objek ke **datetime**

```python
# Merubah data variabel year, 0 = 2011 dan 1 = 2012

df.loc[df["year"] == 0, "year"] = 2011
df.loc[df["year"] == 1, "year"] = 2012
```

```python
df.info()
```

```python
# Merubah data season untuk memudahkan analisis
df.loc[df["season"] == 1, "season"] = "Springer"
df.loc[df["season"] == 2, "season"] = "Summer"
df.loc[df["season"] == 3, "season"] = "Fall"
df.loc[df["season"] == 4, "season"] = "Winter"
```

```python
# Cek data terbaru
df
```

```python
# jika hari bukan akhir pekan atau hari libur adalah 1, sebaliknya adalah 0 (hari libur/akhir pekan)

df.loc[df["workingday"] == 0, "workingday"] = "Holiday"
df.loc[df["workingday"] == 1, "workingday"] = "Working"


df.loc[df["holiday"] == 0, "holiday"] = "Yes"
df.loc[df["holiday"] == 1, "holiday"] = "No"
```

```python
# Ubah data month
df.loc[df["month"] == 1, "month"] = "January"
df.loc[df["month"] == 2, "month"] = "February"
df.loc[df["month"] == 3, "month"] = "March"
df.loc[df["month"] == 4, "month"] = "April"
df.loc[df["month"] == 5, "month"] = "Mei"
df.loc[df["month"] == 6, "month"] = "June"
df.loc[df["month"] == 7, "month"] = "July"
df.loc[df["month"] == 8, "month"] = "August"
df.loc[df["month"] == 9, "month"] = "September"
df.loc[df["month"] == 10, "month"] = "October"
df.loc[df["month"] == 11, "month"] = "November"
df.loc[df["month"] == 12, "month"] = "December"
```

```python
# Handle Missing Value

df.isna().sum()
```

```python
# Mengisi data missing value pada hour dengan mean
df.hour.fillna(value=df.hour.mean(), inplace=True)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Musim apa yang paling tinggi dan rendah dalam menggunakan sistem Bike Sharing dari pengguna casual maupun registered?

```python
df.sample(10)
```

```python
df['year'].unique()
```

```python
df['casual'].sum()
```

```python
casual_season = df.groupby("season").casual.sum().sort_values(ascending=False).reset_index()
casual_season
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(25, 7))

colors = [ "#04364A","#176B87","#64CCC5",  "#DAFFFB", "#D3D3D3"]


sns.barplot(x="casual", y="season", data=casual_season, palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Most Performing Season", loc="center", fontsize=18)
ax[0].tick_params(axis ='y', labelsize=15)

sns.barplot(x="casual", y="season", data=casual_season.sort_values(by="casual", ascending=True), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Lowest Performing Season", loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=15)

plt.suptitle("Most and Lowest Performing Season by Number Casual", fontsize=30)
plt.show()
```

```python
df['registered'].sum()
```

```python
registered_season = df.groupby("season").registered.sum().sort_values(ascending=False).reset_index()
registered_season
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(25, 7))

colors = [ "#04364A","#176B87","#64CCC5",  "#DAFFFB", "#D3D3D3"]


sns.barplot(x="registered", y="season", data=registered_season, palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Most Performing Season", loc="center", fontsize=18)
ax[0].tick_params(axis ='y', labelsize=15)

sns.barplot(x="registered", y="season", data=registered_season.sort_values(by="registered", ascending=True), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Lowet Performing Season", loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=15)

plt.suptitle("Most and Lowest From Perfoming Season by Number Registered", fontsize=30)
plt.show()
```

### Pertanyaan 2: Bagaimana data - data yang ada dari penggunaan mengenai sistem Bike Sharing?

Berdasarkan Season

```python
byseason_df = df.groupby(by="season").instant.nunique().reset_index()
byseason_df.rename(columns={
    "instant": "count"
}, inplace=True)
byseason_df
```

```python
plt.figure(figsize=(10, 5))
colors_season = ["#04364A","#176B87", "#64CCC5" , "#DAFFFB", "#D3D3D3"]
sns.barplot(
    y="count",
    x="season",
    data=byseason_df.sort_values(by="count", ascending=False),
    palette=colors_season
)
plt.title("Count of Season by Bike Sharing Dataset", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

```python
byweathersit_df = df.groupby(by="weathersit").instant.nunique().reset_index()
byweathersit_df.rename(columns={
    "instant": "count"
}, inplace=True)
byweathersit_df
```
